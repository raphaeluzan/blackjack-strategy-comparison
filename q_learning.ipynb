```python
from random import randint
import random
import numpy as np

class CasinoGame(object):
    """
        CasinoGame.
    """

    def __init__(self, max_repetition, max_value):
        # @nb Number of stick to play with
        super(CasinoGame, self).__init__()
        self.MAX_REPETITION = max_repetition
        self.MAX_VALUE = max_value
        self.nb_current = -1
        self.rand = random.randint(0, max_value)
        self.count = 0
        
    def getRand(self):
        return self.rand

    def is_finished(self):
        if self.nb_current == self.rand or self.count==self.MAX_REPETITION:
            return True
        return False

    def reset(self):
        # Reset the state of the game
        self.nb_current = -1
        self.rand = random.randint(0, self.MAX_VALUE)
        self.count = 0
        return self.MAX_REPETITION, self.MAX_VALUE

    def display(self):
        if self.nb_current == self.rand:
            ress = "Bravo vous avez trouve le reponse {} en {} essaies".format(self.rand,self.nb_current)
        elif self.nb_current <= self.rand:
            ress = "superieur a {}".format(self.nb_current)
        else:
            ress = "inferieur a {}".format(self.nb_current)
        print ("essai : {}/{} | {}".format(self.count,self.MAX_REPETITION,ress))

    def step(self, action):
        # @action either 1, 2 or 3. Take an action into the environement
        # @action either [0, self.MAX_VALUE]
        self.count += 1
        self.nb_current = action
        if action == self.rand:
            return None,1
        elif action <= self.rand:
            return self.nb_current, 0
        elif action >= self.rand:
            return self.nb_current, 0
```
